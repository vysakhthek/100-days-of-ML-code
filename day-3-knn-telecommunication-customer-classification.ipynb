```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv('data/teleCust1000t.csv')
df.head()
```

```python
df['custcat'].value_counts()
```

```python
df.columns
```

```python
X = df[['region', 'tenure','age', 'marital', 'address', 'income', 'ed', 'employ','retire', 'gender', 'reside']].values  #.astype(float)
X[0:5]
```

```python
y = df['custcat'].values
y[0:5]
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 4
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

```python
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python
from sklearn.neighbors import KNeighborsClassifier

k = 6
neighbors = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
neighbors
```

```python
y_pred = neighbors.predict(X_test)
y_pred[0:5]
```

```python
from sklearn import metrics

print("Train Accuraccy: ", metrics.accuracy_score(y_train, neighbors.predict(X_train)))
print("Test Accuraccy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
Ks = 10
accuracy = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    accuracy[n-1] = metrics.accuracy_score(y_test, yhat)

accuracy
```

```python
plt.plot(range(1,Ks),accuracy)
# plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
# plt.fill_between(range(1,Ks),mean_acc - 3 * std_acc,mean_acc + 3 * std_acc, alpha=0.10,color="green")
# plt.legend(('Accuracy ', '+/- 1xstd','+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.show()
```
