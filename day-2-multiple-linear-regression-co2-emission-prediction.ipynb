```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/FuelConsumption.csv')
df.head()
```

```python
df.describe()
```

```python
feature_df = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']]
feature_df.head()
```

```python
from sklearn.model_selection import train_test_split


x = feature_df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
y = feature_df[['CO2EMISSIONS']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# print(x_train.count())
# print(x_test.count())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print('Coefficients: ', model.coef_)
```

```python
#Prediction

y_pred = model.predict(x_test)
print('Mean Squared Error (MSE): ', np.mean((y_pred -  y_test)**2))
print('Varience score : ', model.score(x_test, y_test))
```

```python
from sklearn.metrics import r2_score

print('R2-score: ', r2_score(y_test, y_pred))
```

```python
x_new = feature_df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY']]
y_new = feature_df[['CO2EMISSIONS']]

x_new_train, x_new_test, y_new_train, y_new_test = train_test_split(x_new, y_new, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_new_train, y_new_train)

print('Coefficients: ', model.coef_)
```

```python
from sklearn.metrics import r2_score

y_new_pred = model.predict(x_new_test)

print('Mean Squared Error: ', np.mean((y_new_pred - y_new_test)**2))
print('Varience Score: ', model.score(x_new_test, y_new_test))
print('R2-score: ', r2_score(y_new_test, y_new_pred))
```
